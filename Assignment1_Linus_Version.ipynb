```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
file_path = "DST_BIL54.csv"
df = pd.read_csv(file_path)

# ---- Inspect column names (uncomment if needed) ----
print(df.head())

# Assume there is a column called 'time' formatted like '2018M01'
# and a column called 'total'

# Convert time column to datetime
df['time'] = pd.to_datetime(df['time'])

# Sort just to be safe
df = df.sort_values('time')

# Create training set (up to 2023-Dec)
train = df[df['time'] <= '2023-12-31'].copy()

# Create time variable x
# x = year + (month-1)/12
train['x'] = train['time'].dt.year + (train['time'].dt.month - 1) / 12

# Extract variable of interest
y = train['total']

# Plot
plt.figure()
plt.plot(train['x'], y)
plt.xlabel('Time (x)')
plt.ylabel('Total registered vehicles')
plt.title('Total Motor Vehicles in Denmark (Training Data)')
plt.show()

# Construct design matrix
X = np.column_stack((np.ones(len(train)), train['x']))

# OLS estimation
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

theta1_hat = beta_hat[0]
theta2_hat = beta_hat[1]

print("Theta1_hat:", theta1_hat)
print("Theta2_hat:", theta2_hat)
```

```python
# Fitted values
y_hat = X @ beta_hat

# Residuals
residuals = y - y_hat
p=len(beta_hat)

n = len(train)
sigma2_hat = (residuals @ residuals) / (n - p)

# Variance-covariance matrix
var_beta_hat = sigma2_hat * np.linalg.inv(X.T @ X)

se_theta1 = np.sqrt(var_beta_hat[0,0])
se_theta2 = np.sqrt(var_beta_hat[1,1])

print("SE Theta1:", se_theta1)
print("SE Theta2:", se_theta2)
```

```python
plt.figure()

# Observations as points
plt.scatter(train['x'], y, label="Observed data")

# Estimated mean (fitted line)
plt.plot(train['x'], y_hat, color='red', label="Estimated mean")

plt.xlabel("Time")
plt.ylabel("Total registered vehicles")
plt.title("Global Linear Trend Model")
plt.legend()

plt.show()
```

```python
for t in range(len(Y_vec)):
    # 1. Get current observation
    x_t = X_mat[t].reshape(-1, 1)  # Make it a column vector (2,1)
    y_t = Y_vec[t]
    
    # 2. Update the R matrix: R_t = R_{t-1} + x_t * x_t^T
    R_t = R_t + (x_t @ x_t.T)
    
    # 3. Calculate the Prediction Error (Innovation)
    # e_t = y_t - x_t^T * theta_{t-1}
    prediction_error = y_t - (x_t.T @ theta_t)
    
    # 4. Update Parameter Vector: theta_t = theta_{t-1} + R_t^-1 * x_t * e_t
    # Note: np.linalg.solve is more numerically stable than inv()
    # Inverse berechnen
    R_inv = np.linalg.inv(R_t)
    print(R_inv)
    update_term = np.linalg.solve(R_t, x_t) * prediction_error
    theta_t = theta_t + update_term.flatten()
    
    # Save the estimate at this time step
    theta_history.append(theta_t.copy())

theta_history = np.array(theta_history)
print(f"Final Recursive Estimates: \nTheta1: {theta_t[0]}\nTheta2: {theta_t[1]}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


X_mat = np.column_stack((np.ones(len(train)), train['x']))
Y_vec = train['total'].values

# 2. Calculate the "Gold Standard" OLS Estimate
# theta_ols = (X^T X)^-1 X^T Y
theta_ols = np.linalg.solve(X_mat.T @ X_mat, X_mat.T @ Y_vec)

def run_rls(R0_value):
    """Function to run RLS with a specific R0 initialization"""
    R_t = np.eye(2) * R0_value
    theta_t = np.zeros((2, 1))
    history = []
    
    for t in range(len(Y_vec)):
        x_t = X_mat[t].reshape(-1, 1)
        y_t = Y_vec[t]
        
        # Update R_t
        R_t = R_t + (x_t @ x_t.T)
        
        # Update Parameters
        prediction_error = y_t - (x_t.T @ theta_t)
        # Using linalg.solve for numerical stability
        gain = np.linalg.solve(R_t, x_t)
        theta_t = theta_t + gain * prediction_error
        
        history.append(theta_t.flatten())
    
    return theta_t.flatten(), np.array(history)

# 3. Run RLS with different initializations
theta_N_orig, history_orig = run_rls(0.1)     # Your original R0
theta_N_small, history_small = run_rls(1e-9)  # Modified small R0

# 4. Print Comparison
print(f"--- Comparison at t=N ---")
print(f"OLS Estimate:        {theta_ols}")
print(f"RLS (R0 = 0.1):      {theta_N_orig}")
print(f"RLS (R0 = 1e-9):     {theta_N_small}")
print(f"\nDifference (OLS - RLS_small): {theta_ols - theta_N_small}")

# 5. Plotting the Convergence
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Subplot 1: Intercept Convergence
ax1.plot(history_orig[:, 0], label='RLS ($R_0=0.1$)', color='blue', alpha=0.6)
ax1.plot(history_small[:, 0], label='RLS ($R_0=10^{-9}$)', color='green', linestyle='--')
ax1.axhline(theta_ols[0], color='red', linestyle=':', label='OLS Target')
ax1.set_title("Convergence of Intercept ($\\theta_1$)")
ax1.legend()
ax1.grid(True)

# Subplot 2: Slope Convergence
ax2.plot(history_orig[:, 1], label='RLS ($R_0=0.1$)', color='blue', alpha=0.6)
ax2.plot(history_small[:, 1], label='RLS ($R_0=10^{-9}$)', color='green', linestyle='--')
ax2.axhline(theta_ols[1], color='red', linestyle=':', label='OLS Target')
ax2.set_title("Convergence of Slope ($\\theta_2$)")
ax2.set_xlabel("Iterations (t)")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def run_rls_forgetting(X, Y, lam, R0_val=0.1):
    n = len(Y)
    theta_t = np.zeros((2, 1))
    R_t = np.eye(2) * R0_val
    history = []
    
    for t in range(n):
        x_t = X[t].reshape(-1, 1)
        y_t = Y[t]
        
        # R_t Update mit  (Vergessensfaktor)
        R_t = lam * R_t + (x_t @ x_t.T)
        
        # Parameter Update
        err = y_t - (x_t.T @ theta_t)
        gain = np.linalg.solve(R_t, x_t)
        theta_t = theta_t + gain * err
        
        history.append(theta_t.flatten())
        
    return np.array(history)

# Berechnung fr beide Lambdas
hist_07 = run_rls_forgetting(X_mat, Y_vec, 0.7)
hist_99 = run_rls_forgetting(X_mat, Y_vec, 0.99)

# Plotting
burn_in = 5 # Wir lassen die ersten 5 Punkte weg (Burn-in)
t_range = range(burn_in, len(Y_vec))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Plot fr Intercept (Theta 1)
ax1.plot(t_range, hist_07[burn_in:, 0], label='$\lambda = 0.7$ (Short memory)', color='orange')
ax1.plot(t_range, hist_99[burn_in:, 0], label='$\lambda = 0.99$ (Long memory)', color='blue')
ax1.set_title("Evolution of Intercept $\hat{\\theta}_1$ with Forgetting")
ax1.legend()
ax1.grid(True)

# Plot fr Slope (Theta 2)
ax2.plot(t_range, hist_07[burn_in:, 1], label='$\lambda = 0.7$', color='orange')
ax2.plot(t_range, hist_99[burn_in:, 1], label='$\lambda = 0.99$', color='blue')
ax2.set_title("Evolution of Slope $\hat{\\theta}_2$ with Forgetting")
ax2.set_xlabel("Time steps (t)")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
